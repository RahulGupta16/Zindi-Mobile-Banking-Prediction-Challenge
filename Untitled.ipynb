```python
train = df.iloc[0:train.shape[0],:]
test = df.iloc[train.shape[0]:,:]
test.reset_index(drop=True,inplace=True)

label = train.Target
train.drop(columns=['Target'],inplace=True)
test.drop(columns=['Target'],inplace=True)
```

```python
label.value_counts()
```

```python
def model(m, split ,train, label, test, random,need=True):
    te = []
    strat = StratifiedKFold(n_splits= split, shuffle=True, random_state = random)
    for tr_index,te_index in strat.split(train,label):
        
        xtrain,xval = train.loc[tr_index,:],train.loc[te_index,:]
        ytrain,yval = label.loc[tr_index], label.loc[te_index]
        
        m.fit(xtrain,ytrain)
        tr_pred = m.predict(xtrain)
        te_pred = m.predict(xval)
        fpr1,tpr1,thr1 = roc_curve(ytrain,tr_pred)
        fpr2,tpr2,thr2 = roc_curve(yval,te_pred)

        print('Training Loss :{}                                    Testing Loss : {}'.format(roc_auc_score(ytrain,tr_pred),roc_auc_score(yval,te_pred)))

  #      print('Training Loss :{}                                    Testing Loss : {}'.format(auc(fpr1,tpr1),auc(fpr2,tpr2)))
        if need :
            te.append(pd.DataFrame(m.predict(test), index = test.index))
            print(te)
    return(te)
    plt.plot(fpr2,tpr2)
    plt.show()
```

```python
final = model(lgb.LGBMClassifier(n_estimators=1000,random_state=2435),5,train,label,test,34615)
```

```python
pred = np.round(np.mean(pd.concat([final[0],final[1],final[2],final[3],final[4]],axis=1),axis=1)).astype('int')
```
